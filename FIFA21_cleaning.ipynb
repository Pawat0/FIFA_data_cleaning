```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('fifa21_raw_data_v2.csv')
```

```python
df.info()
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
#change column name to lower case
df.columns = map(str.lower, df.columns)
```

```python
#change 'joined' column to datetime data type and create column into 'month', 'day' and 'year' column
df['joined'] = pd.to_datetime(df['joined'])

df['day_joined'] = df['joined'].dt.day
df['month_joined'] = df['joined'].dt.month
df['year_joined'] = df['joined'].dt.year
```

```python
#remove some string and space
df['club'] = df['club'].str.replace('\n\n\n\n', '', regex = True)
df['value'] = df['value'].str.replace('€', '', regex=True)
df['wage'] = df['wage'].str.replace('€', '', regex=True)
df['release clause'] = df['release clause'].str.replace('€', '', regex=True)
df['height'] = df['height'].str.replace('cm', '', regex=True)
```

```python
#clean weight column and change data type to numeric
df['weight_unit'] = np.where(df['weight'].str.contains('lbs'), 'lbs', 'kg')

df['weight'] = df['weight'].str.replace('kg', '', regex=True)
df['weight'] = df['weight'].str.replace('lbs', '', regex=True)

df['weight'] = pd.to_numeric(df['weight'])
```

```python
#clean 'height' column and change feet into cm 
df['height_unit'] = np.where(df['height'].str.contains("'"), 'feet', 'cms')

df['height'] = df['height'].str.replace('cm', '', regex=True)
df['height'] = df['height'].str.replace('"', '', regex=True)

df[['feet', 'inches']] = df['height'].str.split("'", expand = True)

df.loc[df['height_unit'] == 'feet', 'height'] = round((pd.to_numeric(df['feet'])*30.48)+(pd.to_numeric(df['inches'])*2.54))

df.loc[df['height_unit'] == 'feet'].head()

df['height'] = pd.to_numeric(df['height'])
```

```python
df['wage_unit'] = np.where(df['wage'].str.contains('M'), 'M', 'K')
df['value_unit'] = np.where(df['value'].str.contains('M'), 'M', 'K')
df['release_clause_unit'] = np.where(df['release clause'].str.contains('M'), 'M', 'K')
```

```python
#remove some sting
df['wage'] = df['wage'].str.replace('M', '', regex=True)
df['wage'] = df['wage'].str.replace('K', '', regex=True)
df['value'] = df['value'].str.replace('M', '', regex=True)
df['value'] = df['value'].str.replace('K', '', regex=True)
df['release clause'] = df['release clause'].str.replace('M', '', regex=True)
df['release clause'] = df['release clause'].str.replace('K', '', regex=True)
```

```python
#change data type to numeric
df['wage'] = pd.to_numeric(df['wage'])
df['value'] = pd.to_numeric(df['value'])
df['release clause'] = pd.to_numeric(df['release clause'])
```

```python
#multiply value by their unit
df.loc[df['wage_unit'] == 'K', 'wage'] = round(df['wage']*1000)
df.loc[df['wage_unit'] == 'M', 'wage'] = round(df['wage']*1000000)
df.loc[df['value_unit'] == 'K', 'value'] = round(df['value']*1000)
df.loc[df['value_unit'] == 'M', 'value'] = round(df['value']*1000000)
df.loc[df['release_clause_unit'] == 'K', 'release clause'] = round(df['release clause']*1000)
df.loc[df['release_clause_unit'] == 'M', 'release clause'] = round(df['release clause']*1000000)
```

```python
#drop 'weight_unit', 'height_unit', 'feet' and 'inches' column
df = df.drop(['weight_unit', 'height_unit', 'feet', 'inches', 'wage_unit', 'value_unit', 'release_clause_unit'], axis = 1)
```
